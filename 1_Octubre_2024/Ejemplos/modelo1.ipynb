# **CONVERSIÓN DE GRADOS CELSIUS A FAHRENHEIT**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

1.   OBTENEMOS LOS DATOS

```python
grados = {
    "Celsius": [-40, -10, 0, 8, 15, 22, 38],
    "Fahrenheit": [-40, 14, 32, 46.4, 59, 71.6, 100.4]
}
datos = pd.DataFrame(grados)
datos
```

SEPARAMOS CARACTERISTICAS (ENTRADAS) Y ETIQUETAS (SALIDA)

```python
X = datos['Celsius']
y = datos['Fahrenheit']
```

2.   DEFINIMOS UN MODELO DE RED NEURONAL

```python
modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), #Capa de entrada - Una neurona (grados Celsius)
    tf.keras.layers.Dense(units=10),
    tf.keras.layers.Dense(units=1) #Capa de salida - Una neurona (grados Fahrenheit)
])
```

ALGUNAS PROPIEDADES DEL MODELO

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

3.  ENTRENAMOS EL MODELO

```python
modelo.fit(X, y, epochs=1000)
```

4.  EVALUAMOS LOS RESULTADOS DEL MODELO ENTRENADO

```python
plt.xlabel("# Épocas")
plt.ylabel("Magnitud de pérdida")
plt.plot(modelo.history.history['loss'])
plt.show()
```

5.  REALICEMOS UNA PREDICCIÓN

```python
celsius = np.array([123, 50, -80, 14, 7])
prediccion = modelo.predict([celsius])
print(np.around(prediccion))
#print(f"{celsius[0]} grados Celsius son {prediccion[0]} grados Fahrenheit.")
```
